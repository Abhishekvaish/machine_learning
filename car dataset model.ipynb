Importing all the required files

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

```python
dataset=pd.read_csv("car.csv")
dataset.head()
dataset.columns[0:6]
```

```python
dataset.replace("?",-99999,inplace=True)
```

 .    
  .  
Creating a one hot encoder


```python
def one_hot_encode(column):
    unq=column.unique() #storing the unique values of a column
    n=column.nunique() #storing the number of unique values of a column
    list=[]      #creating a list that will hold the converted column
    for i in column:  #taking each value in the column
        app=[]        
        for k in range(n):#making a list [0,0,0.....,upton n]
            app=app+[0]
        for j in range(len(unq)):# accesing the position of each element in unq
            if i==unq[j]:        # if the element at position j of unq is same as the element in column
                app[j]=1         #then add a 1 in that location  
                list.append(app)
                break
    return list
```

```python
#converting all required column into one hot encoding

buying=one_hot_encode(dataset["buying"])
maintainance=one_hot_encode(dataset["maintainance"])
doors=one_hot_encode(dataset["doors"])
person=one_hot_encode(dataset["person"])
lug_boot=one_hot_encode(dataset["lug_boot"])
safety=one_hot_encode(dataset["safety"])
acceptibility=one_hot_encode(dataset["acceptibility"])

print(len(buying),len(maintainance),len(doors),len(person),len(lug_boot),len(safety),len(acceptibility))
```

```python
# storing the converted dataset and outcome  in mdataset and final respectively
mdataset=[]
final=[]
for i in range (1727):
    mdataset.append(buying[i]+maintainance[i]+doors[i]+person[i]+lug_boot[i]+safety[i])
    final.append(acceptibility[i])
```

```python
#convertin into an array
mdataset=np.array(mdataset)
mfinal=np.array(final)
```

```python
#spliting for training and testing
x_train,x_test,y_train,y_test=train_test_split(mdataset,mfinal,test_size=0.2,random_state=35)
```

```python
#building the model,feeding dataset to the model,training,testing...
model=Sequential()
model.add(Dense(500,input_dim=21,activation="sigmoid"))
model.add(Dense(100,activation="sigmoid"))
model.add(Dense(4,activation="softmax"))
model.compile(loss="mean_squared_error",optimizer='adam',metrics=["accuracy"])
model.fit(x_train,y_train,epochs=500,batch_size=50,validation_data=(x_test,y_test))
```

# Frontend python-tkinter

```python
from tkinter import *
from tkinter import messagebox

window=Tk()
window.title("car prediction")
window.geometry("200x290+100+100")
window.resizable(width=False,height=False)

#adding all the labels
ro=1
for i in dataset.columns[0:6]:
    Label(window,text=(i+"  :"),padx=3,pady=1,bd=8,font=("arial",10,"italic")).grid(row=ro,column=0,sticky=E)
    ro=ro+1

############################optionlists###################################
def optionlist(r,c,opt_list):
    fopt=StringVar()
    fopt.set(opt_list[0])
    OptionMenu(window,fopt,*opt_list).grid(row=r,column=c,sticky=E)
    return fopt

###########converting the selected options into 1 and 0 ##################### 
def get_1_0(all_selected_opt):
    inp=[] #to store the input
    for i in range(len(all_selected_opt)):
        temp=[]     #for storing the value of option
        column=dataset.columns #storing names of all column  in dataset
        unq=dataset[column[i]].unique()  #storing the unique values of a column i 
        n=dataset[column[i]].nunique()   #storing number of unique values in column i
        
        for k in range(n):       #making a list [0,0,0.....,upto n]
            temp=temp+[0]
        
        for j in range (len(unq)):   # taking each unique values of  column i and comparing with selected option  
            if all_selected_opt[i]==unq[j]: # if the values are same then making the value of temp at location j 1
                temp[j]=1
                inp=inp+temp  #adding it to the input list
    return inp
                
##############################prediction#################################                
def predict():
    inp=get_1_0([buying.get(),maintainance.get(),doors.get(),person.get(),lug_boot.get(),safety.get()])
    print(len(inp))
    result=model.predict_classes([[inp]])
    if result == 0:
        messagebox.showwarning("Accceptibility"," unacceptale")
    elif result==1:
        messagebox.showinfo("Accceptibility"," acceptale")
    elif result==2:
        messagebox.showinfo("Accceptibility"," very good")
    elif result==3:
        messagebox.showinfo("Accceptibility"," good")
        
        
##########################################################################

buying=optionlist(1,1,['vhigh', 'high', 'med', 'low'])
maintainance=optionlist(2,1,['vhigh', 'high', 'med', 'low'])
doors=optionlist(3,1,['2', '3', '4', '5more'])
person=optionlist(4,1,['2', '4', 'more'])
lug_boot=optionlist(5,1,['small', 'med', 'big'])
safety=optionlist(6,1,['med', 'high', 'low'])

Label(window,text=" ").grid(row=7,column=0)

Button(window,text="predict",padx=8,pady=1,bd=8,font=("arial",12,"bold"),command=predict).grid(row=9,column=0,columnspan=2)

window.mainloop()
```
